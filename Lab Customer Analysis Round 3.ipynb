```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el conjunto de datos

df= pd.read_csv(r"C:\Users\danie\Downloads\marketing_customer_analysis.csv")
```

```python
df
```

```python
print("\nDescripción del DataFrame:")
print(df.describe())
```

```python
# Mostrar un gráfico del número total de respuestas
plt.figure(figsize=(8, 6))
sns.countplot(x='Response', data=df)
plt.title('Número Total de Respuestas')
plt.show()
```

```python
# Mostrar un gráfico de la tasa de respuesta por canal de ventas
plt.figure(figsize=(8, 6))
sns.countplot(x='Sales Channel', hue='Response', data=df)
plt.title('Tasa de Respuesta por Canal de Ventas')
plt.legend(title='Respuesta')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(y='Total Claim Amount', x='Response', data=df)
plt.title('Tasa de Respuesta por Monto Total del Reclamo')
plt.show()
```

```python
# Mostrar un gráfico de la tasa de respuesta por ingreso
plt.figure(figsize=(10, 6))
sns.boxplot(y='Income', x='Response', data=df)
plt.title('Tasa de Respuesta por Ingreso')
plt.show()
```
